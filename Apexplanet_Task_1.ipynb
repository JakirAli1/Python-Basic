# 3. DATA CLEANING & TRANSFORMATION


```python
#  Remove duplicate rows

df = df.drop_duplicates()
```

```python
df
```

```python
# Fill missing values

for col in df.columns:
  if df[col].dtype == 'object':
    df[col] = df[col].fillna(df[col].mode([0]))
  else:
    df[col]=df[col].fillna(df[col].median())
```

```python
# Convert date column to proper format

if "Date_of_Birth" in df.columns:
    df["Date_of_Birth"] = pd.to_datetime(df["Date_of_Birth"], errors='coerce')
```

```python
# Feature Engineering Example: Create Age

if "Date_of_Birth" in df.columns:
    df["Age"] = (pd.Timestamp("today") - df["Date_of_Birth"]).dt.days
```

```python
# Standardize text (example: CustomerName)

if "CustomerName" in df.columns:
    df["CustomerName"] = df["CustomerName"].str.title().str.strip()
```

```python
# Convert columns to lowercase for consistency

df.columns = df.columns.str.lower().str.replace(" ", "_")
```
