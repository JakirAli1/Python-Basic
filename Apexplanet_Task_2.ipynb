# Multivariate Analysis & Correlation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
sns.scatterplot(x='petal_length', y='sepal_width', data=df)
plt.title('sepal_width vs petal_width')
```

```python
# Correlation Matrix or HeatMap
```

```python
corr = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr()
```

```python
sns.heatmap(corr, annot=True) #annotation
```

```python
car_data = {
    'Mileage_k': [10, 20, 30, 40, 50, 60, 70, 80], # in thousands
    'Price_k': [25, 22, 18, 15, 12, 10, 8, 7],    # in thousands
    'Age_Yrs': [1, 2, 3, 4, 5, 6, 7, 8],
    'Brand': ['Toyota', 'Honda', 'Toyota', 'BMW', 'Honda', 'Toyota', 'BMW', 'Honda']
}
df_cars = pd.DataFrame(car_data)
```

```python
#Q1. What is the relation between Mileage_k and price_k using a scatter plot
plt.scatter(df_cars['Mileage_k'], df_cars['Price_k'])
```

```python
#Q2. What is the correlation value between Mileage_k and price_k ?
```

```python
df_cars[['Mileage_k', 'Price_k']].corr()
```

```python
sns.boxplot(x='Price_k', y='Brand', data=df_cars)
```
